```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 8*np.pi, 1000)
```

```python
w1 = 1
w2 = 0.75
x12 = np.cos(w1*t - np.pi/2)
y12 = np.cos(w2*t - np.pi/2)
w3 = 1
w4 = 2
x34 = np.cos(w3*t - np.pi/2)
y34 = np.cos(w4*t - np.pi/2)
x56 = np.cos(w3*t - np.pi/2)
y56 = np.cos(w4*t - 5*np.pi/6)  
x78 = np.cos(w3*t - np.pi/2)
y78 = np.cos(w4*t)
```

```python
fig, ax = plt.subplots(2,2, figsize = (10, 10))
ax[(0,0)].set_title("Lissajous Curves w1 = 3/4 w2 Delta = 0")
ax[(0,1)].set_title("Lissajous Curves w1 = 2w2 Delta = 0")
ax[(0,0)].plot(x12, y12)
ax[(0,1)].plot(x34, y34)
ax[(1,0)].set_title("Lissajous Curves w1 = 2w2 Delta = pi/3")
ax[(1,1)].set_title("Lissajous Curves w1 = 2w2 Delta = pi/2")
ax[(1,0)].plot(x56, y56)
ax[(1,1)].plot(x78, y78)
```
