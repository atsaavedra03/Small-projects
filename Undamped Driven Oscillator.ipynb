```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
```

```python
#Model of Undamped Driven Oscillator

t = np.linspace(0,16*np.pi,1000)
w0 = 2
w = 1.99999999
F = 1
def func(t):
    results = (F/(w0**2 - w**2)) * (np.sin(w*t) - (w/w0)*np.sin(w0*t))
    return results
values = func(t)
plt.plot(t, values)
```
